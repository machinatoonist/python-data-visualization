# Heatmap

```python
sns.set_style('whitegrid')
```

```python
fuel_class = pd.crosstab(df['class_summary'],
                        df['fuel_type_summary'],
                        values = df['comb08'],
                        aggfunc = 'mean')
```

```python
fuel_class.head()
```

```python
sns.heatmap(fuel_class, 
            annot = True, 
            fmt = '.0f')
```

```python
sns.heatmap(fuel_class, 
            annot = True, 
            fmt = '.0f', 
            cmap = 'YlGnBu')
```

```python
fig, ax = plt.subplots(figsize = (8,5))
sns.heatmap(fuel_class, 
            annot = True, 
            fmt = '.0f', 
            cmap = 'icefire', 
            ax = ax)
ax.set(ylabel = 'Car Type',
       xlabel = 'Fuel Type')
fig.savefig('seaborn_heatmap.svg')
```
