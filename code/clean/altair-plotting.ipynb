```python
# Altair data types
# quantitative Q
# ordinal O - discrete ordered quantity eg: 1,2,3,4
# nominal N - discrete unordered category eg: Red, Yellow, Green
# temporal T - time or date value 
# geojson G - geographic shape eg: shape file
```

# Concatenating charts

```python
chart1 = alt.Chart(df).mark_tick().encode(
    x = 'barrels08',
    y = 'fuel_type_summary'
)
chart2 = alt.Chart(df).mark_bar().encode(
    alt.X('barrels08:Q', bin = True),
    alt.Y('count()')
)
chart1 | chart2
```

```python
chart2 & chart1
```

```python
alt.vconcat(chart1, chart2)
```

```python
alt.hconcat(chart1, chart2)
```

# Layers and Annotation

```python
bars = alt.Chart(df).mark_bar().encode(
    x = 'mean(fuelCost08):Q',
    y = 'year:O')

bars
```

```python
rule = alt.Chart(df).mark_rule(color = 'red').encode(
    x = 'mean(fuelCost08):Q'
)

bars + rule
```

```python
text = bars.mark_text(
    align = 'left', dx = 3).encode(text = alt.Text('mean(fuelCost08):Q', format = ',.0f'))

(bars + rule + text)
```

```python
(bars + rule + text).properties(width = 700)
```

```python
linetext = rule.mark_text(
    align = 'left', dx = 3, dy = 200).encode(text = alt.Text('mean(fuelCost08):Q', format = ',.0f'))

(bars + rule + text + linetext).properties(width = 700)
```

# Amazon dataset

```python
src_file = Path.home() / 'git' / 'python-data-visualization' / 'code' / 'data' / 'raw' / 'AmazonBooks.xlsx'
```

```python
src_file
```

```python
df = pd.read_excel(src_file)
df.head()
```

```python
alt.Chart(df).mark_rect().encode(
    x = 'Year:O', 
    y = 'Genre:O',
    color = 'mean(Price):Q',
    tooltip=[alt.Tooltip('mean(Price):Q', format = '$.2f'),
             alt.Tooltip('count(Name):Q', format = '.0f')]
)
```

```python
alt.Chart(df).mark_bar().encode(
    alt.Y('Year:O', title = 'Published Year'),
    alt.X('sum(Reviews)', title = 'Number of Reviews'),
    color = 'Genre')
```

```python
# Small dataset so no need for data server
df.info()
```

```python
top_authors = list(
    df.groupby(['Author'], as_index = False)
      .agg({'Reviews': 'sum'})
      .nlargest(20, columns = ['Reviews'])['Author']
)

top_authors
```

```python
(alt.Chart(df.query('Author == @top_authors'))
    .mark_circle(
        opacity = 0.8,
        stroke = 'black',
        strokeWidth = 1)
    .encode(
        alt.Y('Author'),
        alt.X('Year:O'),
        alt.Size('sum(Reviews)', 
                 scale = alt.Scale(range = [0, 500]),
                 legend = alt.Legend(title = 'Reviews')),
        alt.Color('Author', legend = None))
    .configure_axis(grid = True)
    .transform_filter(
        alt.FieldOneOfPredicate(
            field = 'Author', oneOf = top_authors))
    .properties(
        width = 550,
        height = 475,
        title = 'Amazon Author Reviews'))
```
